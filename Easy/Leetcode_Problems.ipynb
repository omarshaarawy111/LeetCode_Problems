```python
# 1. Two Sum
class Solution():   
    def twoSum(self , nums , target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        nums_dict = {}
        result = []
        
        if not 2 <= len(nums) <= pow(10,4) :
            return False

        for i in range(0,len(nums)) :
            if not  pow(-10,9) <= nums[i] <= pow(10,9) :
                return False

        if not pow(-10,9) <= target <= pow(10,9) :
                return False      

        #O(n)
        for i,num in enumerate(nums) :
            complement = target - num 
            if complement in nums_dict :
                result.append(i)
                result.append(nums_dict[complement])
                return result
            nums_dict[num] = i 

        return result   
```

```python
# 9. Palindrome Number
class Solution(object):   
    def isPalindrome(self, x):
        """
        :type strs: str
        :rtype: bool
        """
        if not pow(-2,31) <= x <= pow(2,31)-1 :
            return False
        if x < 0 :
            return False
        
        return str(x) == str(x)[::-1]
```

```python
# 14. Longest Common Prefix
class Solution(object):
    def longestCommonPrefix(self, strs):
        """
        :type strs: List[str]
        :rtype: str
        """
        if not 1 <= len(strs) <= 200 :
            return False
        
        common_prefix = ""  

        for i in range(len(strs)) :  
           if not 0 <= len(strs[i]) <= 200:
            return False
           
           if i == 0 :
            common_prefix += strs[i]
            continue
           
           temp_common_prefix = "" 
           for char in range(len(strs[i])):
            if strs[i][char].isupper():
                return False  
            
            if char == 0 and len(common_prefix) > 0 :
                if strs[i][0] != common_prefix[0]:
                    common_prefix = ""
                    return common_prefix
                
            elif char == 0 and len(common_prefix) == 0 :
                    return common_prefix
            
            if len(common_prefix) > char :
                if strs[i][char] ==  common_prefix[char]:
                    temp_common_prefix += strs[i][char]
                else :
                    common_prefix = temp_common_prefix
                    break
                
            else :
                break    
           common_prefix = temp_common_prefix 
        return common_prefix
```

```python
# 20. Valid Parentheses
class Solution(object):
    def isValid(self, s):
        """
        :type s: str
        :rtype: bool
        """
        stack = []
        mapping = {')': '(', '}': '{', ']': '['}

        for ch in s:
            if ch in mapping.values():  
                stack.append(ch)
            elif ch in mapping:  
                if not stack or stack[-1] != mapping[ch]:
                    return False
                stack.pop()
            else:
                return False  

        return len(stack) == 0
```

```python
# 21. Merge Two Sorted Lists
# Linked list method
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Helper(object):
    def build_linked_list(self, list):
        dummy = ListNode()
        tail = dummy
        for v in list:
            tail.next = ListNode(v)
            tail = tail.next
        return dummy.next

    def linked_list_to_list(self, node):
        result = []
        while node:
            result.append(node.val)
            node = node.next
        return result
    
class Solution(object):
    def mergeTwoLists(self, list1, list2):
        """
        :type list1: Optional[ListNode]
        :type list2: Optional[ListNode]
        :rtype: Optional[ListNode]
        """
        dummy = ListNode()
        tail = dummy

        while list1 and list2:
            if list1.val <= list2.val:
                tail.next = list1
                list1 = list1.next

            else:
                tail.next = list2
                list2 = list2.next
            tail = tail.next

        if list1:
            tail.next = list1
            
        else:
            tail.next = list2

        return dummy.next
```

```python
# 121. Best Time to Buy and Sell Stock
class Solution(object):
    def maxProfit(self, prices):
        """
        :type prices: List[int]
        :rtype: int
        """
        if not 1 <= len(prices) <= pow(10,5):
            return False
        
        for i in prices :
            if not 0 <= i <= pow(10,4):
                return False
            
        if not len(prices) > 1 :
            return 0
        
        max_profit = 0
        min_price = float("inf")

        for price in prices:
            if price < min_price :
                min_price = price
                
            elif price - min_price > max_profit:
                max_profit = price - min_price   

        return max_profit        
```

```python
# 136. Single Number 
class Solution(object):
    def singleNumber(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """

        if not 1 <= len(nums) <= (3* 10**14):
            return False
        
        for i in nums :
            if not (-3* 10**14) <= i <= (3* 10**14):
                return False

        counter = {}
        for i in nums:
            if i not in counter:
                counter[i] = 1
                
            else :
                counter[i] += 1                    

        for key,value in counter.items():
            if value == 1:
                return key
                
        return False
```

```python
# 643. Maximum Average Subarray I
class Solution(object):
    def findMaxAverage(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: float
        """
        n = len(nums)

        if not 1 <= k <= n <= 10**5 :
            return False
        
        for i in range(len(nums)) :
            if not -10**4 <= nums[i] <= 10**4:
                return False

        current = sum(nums[:k])
        max = current
        for element in range(k, n):
            current += nums[element] - nums[element-k]
            if current > max :
                max = current

        return max / float(k)                
```

```python
# 217. Contains Duplicate
class Solution(object):
    def containsDuplicate(self, nums):
        """
        :type nums: List[int]
        :rtype: bool
        """
        if not 1 <= len(nums) <= 10**5 :
            return False

        for i in nums :
            if not -10**9 <= i <= 10**9:
                return False

        repeated = set()       
        for i in nums:
            if i in repeated :
                return True
            
            else :
                repeated.add(i)        
        return False 
```

```python
# 268. Missing Number
class Solution(object):
    def missingNumber(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        n = len(nums)

        if not 1 <= n <= 10**4:
            return False
        
        for i in nums :
            if not 0 <= i <=n:
                return False
            
        if not n == len(set(nums)):
            return False
        
        for i in range(len(nums)+1) :
            if not i in nums:
                return i         
```

```python
# 283. Move Zeroes
class Solution(object):
    def moveZeroes(self, nums):
        """
        :type nums: List[int]
        :rtype: None Do not return anything, modify nums in-place instead.
        """
        if not 1 <= len(nums) <= 10 ** 4:
            return False
        
        for i in nums :
            if not pow(-2,31) <= i <= pow(2,31) - 1:
                return False
            
        last_non_zero = 0 
        for i in range(len(nums)):
            if nums[i] != 0:
                nums[last_non_zero], nums[i] = nums[i], nums[last_non_zero]
                last_non_zero += 1
                
        return nums        
```

```python
# 344. Reverse String
class Solution(object):
    def reverseString(self, s):
        """
        :type s: List[str]
        :rtype: None Do not return anything, modify s in-place instead.
        """
        if not 1 <= len(s) <= 10 ** 5:
            return False

        for char in s:
            if not char.isascii() and char.isprintable():
                return False    
            
        left, right = 0, len(s)-1
        while right > left:
            s[right], s[left] = s[left], s[right]
            left += 1
            right -=1
            
        return s       
```

```python
# 387. First Unique Character in a String
class Solution(object):
    def firstUniqChar(self, s):
        """
        :type s: str
        :rtype: int
        """
        if not 1 <= len(s) <= 10 ** 5:
            return False
        
        for char in s:
            if not char.islower():
                return False

        for char in s:  
            if not char in s[s.index(char) + 1:]:
                return s.index(char)
          
        return -1
```

```python
# 242. Valid Anagram
class Solution(object):
    def isAnagram(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: bool
        """
        if not (1 <= len(s) <= 5 * 10 ** 4) and (1 <= len(t) <= 5 * 10 ** 4):
            return False

        for char in s:
            if not char.islower():
                return False
        
        for char in t:
            if not char.islower():
                return False
            
        if not len (s) == len(t):
            return False
        
        return set(s) == set(t)
```

```python
# 66. Plus One
class Solution(object):
    def plusOne(self, digits):
        """
        :type digits: List[int]
        :rtype: List[int]
        """
        if not 1 <= len(digits) <= 100:
            return False
        
        for i in digits:
            if not 0 <= i <= 9 :
                return False

        for i in digits:
            i = str(i)
            if len(i) > 1 and i[0] == '0': 
                return False

        lst_to_str = ""
        for i in digits:
            lst_to_str += str(i)   
            
        sum = int(lst_to_str) + 1    

        new_list= []
        while sum:
            new_list.append(sum%10)
            sum //= 10   

        return new_list[::-1]
```

```python
# 26. Remove Duplicates from Sorted Array
class Solution(object):
    def removeDuplicates(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        if not 1 <= len(nums) <= 3 * 10 ** 4:
            return False
        
        for i in nums:
            if not -100 <= i <= 100:
                return False
            
        k = 1
        for i in range(1,len(nums)):
            if nums[i] != nums[i-1]:
                nums[k] = nums[i]
                k += 1

        return k
```

```python
# 27. Remove Element
class Solution(object):
    def removeElement(self, nums, val):
        """
        :type nums: List[int]
        :rtype: int
        """
        if not 0 <= len(nums) <= 100:
            return False
        
        for i in nums:
            if not 0 <= i <= 50:
                return False

        if not 0 <= val <= 50:
                return False
            
        k = 0
        for i in range(1,len(nums)):
            if nums[i-1] != val:
                k += 1
                
        return k
```

```python
# 1523. Count Odd Numbers in an Interval Range
class Solution(object):
    def countOdds(self, low, high):
        """
        :type low: int
        :type high: int
        :rtype: int
        """
        if not 0 <= low <= high <= 10 ** 9:
            return False
        
        total_count = high - low + 1

        odd_count = total_count // 2

        if total_count %2 != 0 and low %2 != 0 :
            odd_count += 1

        return odd_count      
```
